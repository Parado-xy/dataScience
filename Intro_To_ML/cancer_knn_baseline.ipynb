Let benchmark a KNNs accuracy as the number of neighbours vary.

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify = cancer.target, random_state = 66
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
training_accuracy = []
test_accuracy = []
# try n_neighbours from 1 to 10
neighbors = range(1,11)

for neighbor in neighbors:
    # build the model
    clf = KNeighborsClassifier(n_neighbors = neighbor)
    # fit the model
    clf.fit(X_train, y_train)
    # Get the train accuracy
    training_accuracy.append(clf.score(X_train, y_train))
    # Get the test accuracy
    test_accuracy.append(clf.score(X_test, y_test))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(neighbors, training_accuracy, label = 'training_accuracy', marker = '*')
plt.plot(neighbors, test_accuracy, label = 'test_accuracy', marker = '*')
plt.ylabel('Accuracy')
plt.xlabel('neighbors')
```

The plot shows the training and test accuracy as the neighbours vary. This model performs best when neighbours = 6
