```python
# df.isnull(). The pandas isnull( ) method returns a dataframe with the same shape sa the original dataframe, but with true or false values
# indicating whether each value in the column is missing or not. Missing values such as NaN or None will be encoded as True, it'll be false  otherwise. 
df.isnull() 
```

```python
#df.fillna(). This method is used to fill in missing values in a DataFrame with a specified value or a method. By defalut, it fills missing slots with NaN, but you can
# specify a different value to use instead. 
# value : scalar, dict, Series, or DataFrame
#    Value to use to fill holes (e.g. 0), alternately a dict/Series/DataFrame of values specifying which value to use for each index (for a Series) or column (for a DataFrame). Values not in the dict/Series/DataFrame will not be filled. This value cannot be a list.
# method : {{'backfill', 'bfill', 'ffill', None}}, default None
#    Method to use for filling holes in reindexed Series:

# ffill: propagate last valid observation forward to next valid.
# backfill / bfill: use next valid observation to fill gap.
# axis : Specifies the axis along which to fill in missing values. It can be 0 (rows), or 1 (columns)
# inplace : bool, default False
#    If True, fill in-place. Note: this will modify any other views on this object (e.g., a no-copy slice for a column in a DataFrame).
# limit : int, default None
#    If method is specified, this is the maximum number of consecutive NaN values to forward/backward fill. In other words, if there is a gap with more than this number of consecutive NaNs, it will only be partially filled. If method is not specified, this is the maximum number of entries along the entire axis where NaNs will be filled. Must be greater than 0 if not None.
# downcast : dict, default is None
#    A dict of item->dtype of what to downcast if possible, or the string 'infer' which will try to downcast to an appropriate equal type (e.g. float64 to int64 if possible).

df.fillna(0) # Fill missing values with zero. 

df.fillna(method= 'ffill') # Forward-fill missing values (propagates last valid observation forward to next).
df.fillna(method='bfill') # Backward-fill missing values (propagates next valid observation backeard to last).
df.interpolate() # Fill missing values via interpolation
```

```python
# It is important to note that the fillna() method returns a new data frame. To work on the same dataframe, set the inplace attribute to True, or reassign the df to a variable 
# of the same name.
```

```python
# df.dropna() this is a method used to remove missing or null values from a dataframe. It removes rows or columns from a dataframe where at least one column is missing. 
# You can remove all the rows containing atleast one missing value by doing df.dropna()
df_drop_na = df.dropna()
```

```python
# If you want to remove only the columns that contain at least one missing value, you can use df.dropna(axis = 1)
df_drop_na = df.dropna(axis = 1)
```

```python
# You can also set a `tresh` value to keep only the rows/columns that have at least `thresh`` non-null values.
df_drop_na = df.dropna(thresh=2)
```

```python
# df.drop() is a method used to remove rows oe columns from a DataFrame by specifying the corresponding labels. It can be used to drop one or multiple
# rows or columns based on their labels. 
# You can remove a specific row by calling df.drop() and passing it the index label of the row you want to remove, and the axis parameter set to 0 (defaults to 0)
df.drop(0)
```

```python
# You can also drop multiple rows by passing a list of index. 
df.drop([0,1])
```

```python
# You can drop columns by passing their labels abd setting the axis parameter to 1
df.drop(['Gender', 'Rank'], axis= 1)
```

```python
# pd.pivot_table(). This is a method used to create a pivot table from a dataframe. A pivot table is a table that summarizes and organizes the data in a more meaningful way, by
# creating a new table with one or more columns as the index, one or more columns as values and one or more columns as attributes. 
# In the example below, we will create a pivot table with Ethnicity as the index and aggregate the sum of the counts. This is used to know the count of each Ethnicity in the
# Dataset.

pivot_table = pd.pivot_table(df, index='Ethnicity', values='Count',aggfunc='sum')
pivot_table.sort_values(by='Count', ascending= False).head() 
```

```python
# You can also include more columns in the pivot table by specifying multiple index and value parameters. You can also include more aggregate functions. 
pivot_table = pd.pivot_table(df, index = ['Ethnicity','Gender'], values = 'Count', aggfunc = ['sum'])
pivot_table.head(20)
```

```python
# df.groupby(). This is a method that is used to group rows of a dataframe based on one or multiple columns. This allows you to perform aggregate operations on the groups, such
# as calculating the mean, median, sum, or count of the values in each group. 
# df.groupby returns a GroupBy object, which you can then use to perform various operations on the groups, such as calculating the sum, mean, or count of the values in each group.
grouped = df.groupby('Gender')
grouped.sum()
# Note that you can get more specific by passing a dict to groupby. for example passing {'Gender':'MALE'} groups only thosee that are male.
```

```python
# df.transpose(). This is a method in the Pandas library used to transpose rows and columns of a DataFrame. This means that the rows become the column and the column become
# rows. 

df_transposed = df.transpose()
df_transposed.head()
```

```python
# Note that using the T attribute of the dataframe will do the same thing as the transpose method. 
df_transposed = df.T
df_transposed.head()
```

```python
# df.merge(). This is a function that allows you to combine two dataframes based on  one or more common columns. The function returns a new DataFrame that contains only the 
# rows where the values in the specified columns match between the two dataframes. Here's an example:

# First dataFrame.
df1 = pd.DataFrame({'key':['A','B','C', 'D'],
                    'value': [1, 2, 3, 4, ]})

df2 = pd.DataFrame({'key':['B','D','E', 'F'],
                    'value': [5, 6, 7, 8, ]})

# Merge the 2 Data|Frames on the 'key' column.

merged_df = df1.merge(df2, on='key')
merged_df.head() # Notice how only the rows their key column had in common merged
```

```python
# df.rename(). This function allows us to change the name of one or more rows/columns in a dataframe. You pass a dict to the column parameter to change column names, and 
# a dict to the index parameter to change index names. 
df_rename = df.rename(columns={'Count':'count', 'Rank':'rank'})
df_rename.head()
```

```python
# you can also rename the index similarly:
df_rename = df.rename(index={0:'first', 1:'second'})
df_rename.head()
```
